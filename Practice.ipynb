```python
# Call dependencies

from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Print/Show html formatted
print(html_soup.prettify())
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Testing quotes in one page, but looking for the second col-md-8 division as there are 2
quotesdiv = html_soup.find_all('div', class_='col-md-8')
quotes = quotesdiv[1].find_all('div', class_='quote')

for q in quotes:
    qtext = q.find('span', class_='text').text
    print(qtext)
```

```python
# Now going directly to search all span tags with class = quote instead as there are no more in the page
quotes = html_soup.find_all('span', class_='text')
for q in quotes:
    print(q.text)
```

```python
# Get all on up to 5 pages
for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('span', class_='text')
    print('Page:', x, '-----------------------------------------------------------')
    for quote in quotes:        
        print(quote.text)
    
    # Go to the next page
    browser.links.find_by_partial_text('Next').click()
```
